```python
import tensorflow as tf
import numpy as np
import os
from matplotlib import pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
```

```python
n = 100
x = np.random.rand(n, 1)
y = np.sin(2 * np.pi * x) + 0.4 * np.random.rand(n, 1)

plt.plot(x, y, '.')
```

```python
# 1 - 100 - 100 - 1

inputs = keras.Input(shape = (1, ))
dense1 = layers.Dense(100, activation = 'relu')
dense2 = layers.Dense(100, activation = 'relu')
dense3 = layers.Dense(1)

output = dense3(dense2(dense1(inputs)))

model = keras.Model(inputs, output)
model.compile(optimizer = keras.optimizers.SGD(learning_rate = 0.1), loss = keras.losses.MeanSquaredError())
```

```python
model.fit(x, y, epochs = 400)
```

```python
z = model.predict(x)

plt.plot(x, y, '.')
plt.plot(x, z, '.r')

plt.show()
```
